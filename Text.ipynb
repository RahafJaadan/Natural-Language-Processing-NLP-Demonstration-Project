# Tokenize

```python
from nltk.tokenize import word_tokenize
```

```python
text="Hello Welcome, to NLP Tutorials.this course will help you to understand Text processing!"
```

```python
word_tokenize(text)
```

```python
import re
cleaned_text = re.sub(r'[^\w\s]', '', text)
print(cleaned_text)
```

\w: Represents any word character, which includes:

- Alphabetical characters (a-z, A-Z)
- Digits (0-9)
- Underscores (_)

\s: Represents any whitespace character

# Stemming Vs Lemmatization

```python
from nltk.stem import PorterStemmer
```

```python
words=["eating","eats","writing","writes","programming","programs","history","finally","finalized"]
```

PorterStemmer

```python
stemming=PorterStemmer()
for word in words:
    print(word+"---->"+stemming.stem(word))
```

Lemmatizer

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```

# Stop Words

```python
text = "The sun rises In the morning and sets in the evening. People enjoy walking in the park and talking with friends."
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words('arabic')
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
words = nltk.word_tokenize(text)
```

```python
words
```

```python
for word in words:
  words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
```
